```python
top_genres = 5
top_genres_df = data_file["Fav genre"].value_counts().nlargest(top_genres)
top_genres_df
Othertop_df = data_file["Fav genre"].value_counts().nsmallest(16 - top_genres).sum()
top_genres_df["Other"]= Othertop_df 
top_genres_df
```

```python
Labels = top_genres_df.index.tolist()

explode = np.zeros(len(top_genres_df))
explode[0] = 0.1
# Count the occurrences of each genre in the dataset
genre_counts = data_file["Fav genre"].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(top_genres_df, labels=Labels, explode= explode, autopct="%1.1f%%", startangle=140)
plt.title("Distribution of Favorite Music Genres")
print (genre_counts)
```

```python
# Group the data by genre and calculate the mean BPM for each group
bpm_by_genre = data_file.groupby('Fav genre')['BPM'].mean()
# Convert the grouped data into a DataFrame
bpm_by_genre_df = pd.DataFrame(bpm_by_genre).reset_index()

# Rename the columns for clarity
bpm_by_genre_df.columns = ['Fav genre', 'Average BPM']

# Display the DataFrame
print(bpm_by_genre_df)
```

```python
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(bpm_by_genre_df['Average BPM'], bins=13, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Average BPM')
plt.ylabel('Frequency')
plt.title('Distribution of Average BPM by Music Genre')

# Define a color palette for the labels
color_palette = ['red', 'green', 'blue', 'orange', 'purple', 'Teal', 'lime', 'pink', 'yellow', 'cyan', 'Magenta', 'Indigo', 'Maroon', 'Lavender', 'Coral', 'Salmon']


# Add labels to each bar
for index, row in bpm_by_genre_df.iterrows():
    plt.text(row['Average BPM'], 1.5, row['Fav genre'], rotation=90, ha='center' , va='bottom', fontsize=min(10, 8), color=color_palette[index]) 

# Adjust the x-axis limits
plt.xlim(min(bpm_by_genre_df['Average BPM']) - 1, max(bpm_by_genre_df['Average BPM']) + 10)

# Show the plot
plt.grid(False)
plt.show()
```

```python
# Define the age groups
age_bins = [0, 18, 25, 35, float('inf')]
age_labels = ['0-17', '18-25', '26-35', '35+']

# Create bins for age groups
data_file['Age Group'] = pd.cut(data_file['Age'], bins=age_bins, labels=age_labels, ordered=True)

# Group the data by age group and music genre, and count occurrences
age_genre_counts = data_file.groupby(['Age Group', 'Fav genre']).size().reset_index(name='Count')

# Convert the grouped data into a DataFrame
age_genre_counts_df = pd.DataFrame(age_genre_counts)

# Display the DataFrame
print(age_genre_counts_df)

# Determine the most frequent genre for each age group
overall_fav_genre = age_genre_counts.loc[age_genre_counts.groupby('Age Group')['Count'].idxmax()]

# Convert the grouped data into a DataFrame
overall_fav_genre_df = pd.DataFrame(overall_fav_genre)

# Display the DataFrame
print(overall_fav_genre_df)
```

```python

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(data_file['Age Group'], bins=4, color='skyblue', edgecolor='black')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Sort the data by count in descending order
age_genre_counts_sorted = age_genre_counts.sort_values(by=['Age Group', 'Count'], ascending=[True, False])

# Get the top 5 genres for each age group
top_5_genres_per_age_group = age_genre_counts_sorted.groupby('Age Group').head(5)

# Plot histograms for each age group
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, (age_group, group_data) in enumerate(top_5_genres_per_age_group.groupby('Age Group')):
    ax = axs[i // 2, i % 2]
    ax.bar(group_data['Fav genre'], group_data['Count'])
    ax.set_title(f'Top 5 Genres for Age Group {age_group}')
    ax.set_xlabel('Genre')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame based on 'Hours per day'
sorted_data = data_file.sort_values(by='Hours per day')

# Select the rows with the lowest and highest values of 'Hours per day'
lowest_hours_data = sorted_data.head(50)  # Assuming 50 individuals spend the least hours per day
highest_hours_data = sorted_data.tail(50)  # Assuming 50 individuals spend the most hours per day

# Calculate the mean of mental health scores for each group
lowest_hours_mean = lowest_hours_data[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean()
highest_hours_mean = highest_hours_data[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean()

# Plot a bar graph to visualize the comparison
plt.figure(figsize=(10, 6))
categories = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
bar_width = 0.35
index = range(len(categories))

plt.bar(index, lowest_hours_mean, bar_width, label='Lowest Hours per day')
plt.bar([i + bar_width for i in index], highest_hours_mean, bar_width, label='Highest Hours per day')

plt.xlabel('Mental Health Condition')
plt.ylabel('Average Score')
plt.title('Comparison of Mental Health Scores for Lowest and Highest Hours per day')
plt.xticks([i + bar_width / 2 for i in index], categories)
plt.legend()
plt.show()
```

```python
# Define the age groups
age_bins = [0, 18, 25, 35, float('inf')]
age_labels = ['0-17', '18-25', '26-35', '35+']

# Create bins for age groups
data_file['Age Group'] = pd.cut(data_file['Age'], bins=age_bins, labels=age_labels)

# Group the data by age group
age_grouped_data = data_file.groupby('Age Group')

# Define categories for mental health conditions
categories = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
num_categories = len(categories)
bar_width = 0.35
index = range(num_categories)

# Plot a bar graph for each age group
for age_group, group_data in age_grouped_data:
    # Sort the data based on 'Hours per day'
    sorted_data = group_data.sort_values(by='Hours per day')
    
    # Select the rows with the lowest and highest values of 'Hours per day'
    lowest_hours_data = sorted_data.head(50)  # Assuming 50 individuals spend the least hours per day
    highest_hours_data = sorted_data.tail(50)  # Assuming 50 individuals spend the most hours per day
    
    # Calculate the mean of mental health scores for each group
    lowest_hours_mean = lowest_hours_data[categories].mean()
    highest_hours_mean = highest_hours_data[categories].mean()
    
    # Plot a bar graph to visualize the comparison
    plt.figure(figsize=(10, 6))
    plt.bar(index, lowest_hours_mean, bar_width, label='Lowest Hours per day')
    plt.bar([i + bar_width for i in index], highest_hours_mean, bar_width, label='Highest Hours per day')
    
    plt.xlabel('Mental Health Condition')
    plt.ylabel('Average Score')
    plt.title(f'Comparison of Mental Health Scores for Lowest and Highest Hours per day (Age Group: {age_group})')
    plt.xticks([i + bar_width / 2 for i in index], categories)
    plt.legend()
    plt.show()
```

```python
# Define the categories of mental health conditions
categories = ['Anxiety','Depression', 'Insomnia', 'OCD']

# Calculate the mean of mental health scores for each age group and hours per day spent
grouped_data = data_file.groupby(['Age Group', 'Hours per day'])[categories].mean().reset_index()

# Plot a scatter plot for each mental health condition
for category in categories:
    plt.figure()
    for age_group, group_data in grouped_data.groupby('Age Group'):
        plt.scatter(group_data['Hours per day'], group_data[category], label=age_group)

    plt.xlabel('Hours per day')
    plt.ylabel(category + ' Score')
    plt.title('Relationship between Hours per day and ' + category + ' Score')
    plt.legend(title='Age Group')
    plt.grid(True)
    plt.show()

    # Calculate the correlation coefficient
    correlation_coefficient = grouped_data.groupby('Age Group')[['Hours per day', category]].corr().iloc[0, 1]
    print(f'Correlation Coefficient for {category}: {correlation_coefficient}')
```
