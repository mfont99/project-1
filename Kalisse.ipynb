```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
cleaned_csv = "https://raw.githubusercontent.com/mfont99/project-1/Matt/cleaned_data.csv"
mm_df = pd.read_csv(cleaned_csv)
mm_df
```

```python
mm_df.describe()

#Anxiety: mean - 5.87; std-2.76; 25% - 4; 50% - 6; 75% - 8
#Depression: mean - 4.88; std - 2.99; 25% - 2; 50% - 5; 75% - 7
#Insomnia: mean - 3.83; std - 3.11; 25% - 1; 50% - 3; 75% - 6
#OCD: mean - 2.66; std - 2.85; 25% - 0; 50% - 2; 75% - 5
#ALL: min/max - 1/10
```

```python
fig = plt.figure(figsize=(11,5))

plt.suptitle("Count of Submitters with a Backgroud in Music")

axs = fig.add_subplot(121)

play = mm_df['Instrumentalist'].value_counts()
play.plot(kind='pie', label = ("Instrumentalist"), colors = ["lightgreen", "lightblue"], autopct='%1.1f%%')

axs = fig.add_subplot(122)

comp = mm_df['Composer'].value_counts()
comp.plot(kind='pie', label = ("Composer"), colors = ["lightgreen", "lightblue"],autopct='%1.1f%%')


```

```python
#Applying a random ID number to each row so that when the Genre columns and the Disorder columns are unpivoted,
#they can be joined back to the original responder via that ID
mm_df["random_id"] = np.random.randint(low=1, high=616, size=len(mm_df))
mm_df.columns
```

```python
#Unpivoting each unique Genre columns into one column called Genre and moving the Frequency/Response to another called "Frequency"

gen_piv = pd.melt(mm_df, id_vars=['random_id'], value_vars=['Classical', 'Country', 'EDM', 'Folk',
       'Gospel', 'Hip Hop', 'Jazz', 'K Pop', 'Latin', 'Lofi', 'Metal', 'Pop',
       'R&B', 'Rap', 'Rock', 'Video Game Music'], var_name='Genre', value_name='Frequency')

gen_piv
```

```python
#Unpivoting each unique mental health columns into one column called Disorder and moving the Severity/Response to another called "Severity"

men_piv = pd.melt(mm_df, id_vars=['random_id'], value_vars=['Anxiety', 'Depression',
       'Insomnia', 'OCD'], var_name='Disorder', value_name='Severity')

men_piv
```

```python
men_piv.groupby(['Disorder']).describe()

men_piv.groupby(['Disorder']).std()
#Dep - 2.99, Anx - 2.76, Ins - 3.11, OCD - 2.85

men_piv.groupby(['Disorder']).mean()
#Dep - 4.88, Anx - 5.87, Ins - 3.83, OCD - 2.66
```

```python
dis_mean = [5.87,4.88,3.83,2.66]
x_axis = np.arange(len(dis_mean))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,["Anxiety", "Depression","Insomnia", "OCD"])


plt.xlabel('Disorders')
plt.ylabel('Severity - Self Determined')
plt.title('Average Severity of Disorders (Self-determined)')

plt.bar(x_axis, dis_mean, color = 'g', alpha=.5, align = "center")
plt.show()
```

```python
# Data for the grouped bar chart
disorders = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
all_avg = [5.87, 4.88,3.83,2.66]
imp_avg = [6.11,4.95,3.84,2.80]
noimp_avg = [5.13, 4.68, 3.77, 2.24]

# Create the figure and axis
fig, ax = plt.subplots()

# Define the number of groups and bar width
num_categories = len(disorders)
bar_width = 0.25
index = np.arange(num_categories)

# Plot the bars for each group
bar1 = ax.bar(index, all_avg, bar_width, label='All', color = "blue")
bar2 = ax.bar(index + bar_width, imp_avg, bar_width, label='Improved', color = "green")
bar3 = ax.bar(index + 2*bar_width, noimp_avg, bar_width, label='No Improvement/Worsened', color = "orange")

# Add labels and title
ax.set_xlabel('Disorders')
ax.set_ylabel('Severity (Self-Reported: Scale 0 to 10)')
ax.set_title('Mean Severity of Disorders by Reported Effect of Music')
ax.set_xticks(index + 2*bar_width / 2)
ax.set_xticklabels(disorders)
ax.legend()

plt.show()
```
