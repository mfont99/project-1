```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

file_path = "https://raw.githubusercontent.com/mfont99/project-1/Matt/cleaned_data.csv"

data = pd.read_csv(file_path)

data.head(10)
```

```python
percentage_improve = (len(data[data["Music effects"] == "Improve"]) / len(data["Music effects"])) * 100

percentage_worsen = (len(data[data["Music effects"] == "Worsen"]) / len(data["Music effects"])) * 100

percentage_no_effect = (len(data[data["Music effects"] == "No effect"]) / len(data["Music effects"])) * 100
```

```python
data["Fav genre"].value_counts()
```

```python
genre_dict = {}
```

```python
complete_data_improve_percentage = (len(data[data["Music effects"] == "Improve"]) / len(data["Music effects"])) * 100
print(f"TOTAL IMPROVE PERCENTAGE: {complete_data_improve_percentage}%\n\n\n")

genres = ["Classical","Country","EDM","Folk","Gospel","Hip Hop","Jazz","K Pop","Latin","Lofi","Metal","Pop",
         "R&B","Rap","Rock","Video Game Music"]

for genre in genres:
    
    current_genre = data[(data[genre] == "Very frequently") | (data[genre] == "Sometimes")]
    
    genre_improve = len(current_genre[current_genre["Music effects"] == "Improve"])
    total_in_genre = len(current_genre["Music effects"])
    percentage_improve = (genre_improve / total_in_genre) * 100
    
    genre_dict[genre] = percentage_improve
    
    print(f"{genre} improve percentage: {percentage_improve}%\n")

genre_dict["Total Average"] = complete_data_improve_percentage    

print(genre_dict)
```

```python
sorted_dict = dict(sorted(genre_dict.items(), key=lambda item: item[1]))
colors = ['blue', 'blue', 'blue', 'blue','red', 'blue', 'blue', 'blue','blue', 'blue', 'blue', 'blue','blue', 'blue', 'blue', 'blue','blue']
x_values = list(sorted_dict.keys())
y_values = list(sorted_dict.values())
plt.xticks(rotation = 90)
plt.xlabel('Genre')
plt.ylabel('Percent that Improve')
plt.title('% Improvement by Genre')

bars = plt.bar(x_values, y_values, color = colors, alpha = 0.60)
    
plt.show()
```

```python
data["Average Mental Health"] = data[['Anxiety', 'Depression', 'Insomnia','OCD']].mean(axis=1)
data
```

```python
m, b, r, p, err = st.linregress(data["BPM"], data["Average Mental Health"])
line_eq = m * data["BPM"] + b

plt.scatter(data["BPM"], data["Average Mental Health"],  marker = 'o', color = 'r')
plt.plot(data["BPM"], line_eq, "g-")
plt.xlabel("Beats per Minute (BPM)")
plt.ylabel("Average Mental Health")
plt.title("Average Mental Health vs BPM")
plt.show()
```

```python
m, b, r, p, err = st.linregress(data["BPM"], data["Age"])
line_eq = m * data["BPM"] + b

plt.scatter(data["BPM"], data["Age"],  marker = 'o', color = 'g')
plt.plot(data["BPM"], line_eq, "r-")
plt.xlabel("Beats per Minute (BPM)")
plt.ylabel("Age")
plt.title("Age vs BPM")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
categories = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
bar_width = 0.10
index = range(len(categories))
colors = ["green","red","brown","blue","orange","purple"]

spotify = data[data["Primary streaming service"] == "Spotify"][["Anxiety","Depression","Insomnia","OCD"]].mean()

yt = data[data["Primary streaming service"] == "YouTube Music"][["Anxiety","Depression","Insomnia","OCD"]].mean()

no_serv = data[data["Primary streaming service"] == "I do not use a streaming service."][["Anxiety","Depression","Insomnia","OCD"]].mean()

apple = data[data["Primary streaming service"] == "Apple Music"][["Anxiety","Depression","Insomnia","OCD"]].mean()

other = data[data["Primary streaming service"] == "Other streaming service"][["Anxiety","Depression","Insomnia","OCD"]].mean()

pandora = data[data["Primary streaming service"] == "Pandora"][["Anxiety","Depression","Insomnia","OCD"]].mean()

plt.bar(index, spotify, bar_width, label='Spotify')
plt.bar([i + bar_width for i in index], yt, bar_width, label='YouTube')
plt.bar([i + 2*bar_width for i in index], no_serv, bar_width, label='None')
plt.bar([i + 3*bar_width for i in index], apple, bar_width, label='Apple')
plt.bar([i + 4*bar_width for i in index], other, bar_width, label='Other')
plt.bar([i + 5*bar_width for i in index], pandora, bar_width, label='Pandora')

plt.xlabel('Mental Health Condition')
plt.ylabel('Average Score')
plt.title('Mental Health vs Music Service')
plt.xticks([i + (5*bar_width) / 2 for i in index], categories)
plt.legend()
plt.show()
```
